## Concatenate

Menempelkan 2 dataframe atau lebih

```python
dfA = pd.DataFrame({
    'Jakarta':[1,2,3,4],
    'Bandung':[5,6,7,8],
    'Padang':[9,10,11,12]
}, index= [0,1,2,3])


dfB = pd.DataFrame({
    'Jakarta':[10,20,30,40],
    'Bandung':[50,60,70,80],
    'Padang':[90,100,110,120]
}, index= [4,5,6,7])


dfC = pd.DataFrame({
    'Denpasar':[100,200,300,400],
    'Merauke':[500,600,700,800],
    'Kupang':[900,1000,1100,1200]
}, index= [8,9,10,11])

display(dfA, dfB, dfC)
```

```python
pd.concat([dfA, dfB])
# defualtnya axis=0 --> menempelkan ke bawah
```

```python
pd.concat([dfA, dfC])
```

```python
pd.concat([dfB, dfC, dfA]) 
```

```python
pd.concat([dfA, dfC], axis=1) 
```

```python
# mengubah label index di dfC
dfC = dfC.rename(index={10:2, 11:3})
dfC
```

```python
display(dfA, dfC)
```

```python
pd.concat([dfA, dfC], axis=0)

# axis=0 
# yg bertambah adalah jumlah baris
# kolom menyesuaikan, 
# kalau ada kolom yg sama, nilai akan lanjut di kolom yg sama
# kalau kolomnya beda semua, maka akan buat kolom baru
```

```python
pd.concat([dfA, dfC], axis=1)

# axis=1 
# yg bertambah adalah jumlah kolom
# baris menyesuaikan, 
# kalau ada label baris yg sama, nilai akan lanjut di label baris yg sama
# kalau label barisnya beda semua, maka akan buat label baris baru
```

```python
# ubah index di dfB
dfB.index = [0,1,2,3]
dfB

# sekarang label index ([0,1,2,3]) dan kolom (Jakarta Bandung Padang) di dfA dan dfB sama
```

```python
pd.concat([dfA, dfB], axis=0) 
```

```python
pd.concat([dfA, dfB], axis=1)
```

```python
# join(how='outer') mirip dengan concat(axis=1)
pd.DataFrame.join(dfA,dfB, how='outer', lsuffix='_a', rsuffix='_b')
```

## Pandas Operation

```python
# buat dictionary untuk datanya
dict_salary = {
    'name':['Kim Jisoo', 'Robert Downey', 'Johny Depp', 'Harry Maguire', 'Ana De Armas'],
    'gender':['F','M','M','M','F'],
    'hire_date':['2018-05-01', '2017-08-01', '2018-11-01', '2019-02-01', '2017-03-01'],
    'salary':[7000, 9000, 8000, 4000, 9000]
}

# buat list berisi nomor karyawan untuk indexnya
emp_number = np.arange(1001,1006)

# membuat dataframe dan disimpan dalam varieble df
df = pd.DataFrame(dict_salary, index=emp_number)
df
```

**Soal:** buat kolom baru bernama 'bonus' berisi 10% dari salary (dalam betuk integer)

```python
# cara 1
my_bonus = df[['salary']] * 0.1
my_bonus = my_bonus.rename(columns={'salary':'bonus'})

my_bonus 
```

```python
df.join(my_bonus)
```

```python
# cara 2: astype untuk mengubah tipe data
df['bonus'] = df['salary'] * 0.1

df['bonus'] = df['bonus'].astype(int)
df
```

```python
# cara 3: floor division
df['bonus'] = df['salary'] // 10
df
```

```python
# cara 4: looping

# looping tiap item dalam kolom salary
# itemnya dibagi 10
# hasilnya dimasukkan ke list kosong
list_bonus = []

for i in df['salary']:
    hasil = i//10
    list_bonus.append(hasil)

list_bonus

# list kosongnya disimpan dalam df['bonus']
df['bonus'] = list_bonus
df 
```

```python
import time
```

```python
start = time.time()
list_bonus = []

for i in df['salary']:
    hasil = i//10
    list_bonus.append(hasil)

list_bonus

# list kosongnya disimpan dalam df['bonus']
df['bonus'] = list_bonus
df 


end = time.time()

print(end-start)
```

```python
# cara 5: list comprehension
df['bonus'] = [i//10 for i in df['salary']]
df 
```

```python
start = time.time()

df['bonus'] = [i//10 for i in df['salary']]
df 

end = time.time()

print(end-start)
```

**apply()**

- Butuh series
- Butuh function (boleh regular function atau lambda)

```python
# butuh series
df['salary']
```

```python
# butuh function
def bagi10(gaji):
    hasil = gaji//10
    return hasil 
```

```python
# mengaplikasikan function bagi10 pada tiap item dalam df['salary']
df['salary'].apply(bagi10) 
```

```python
df['bonus_apply'] = df['salary'].apply(bagi10) 
df
```

```python
start = time.time()

# butuh function
def bagi10(gaji):
    hasil = gaji//10
    return hasil 

df['bonus_apply'] = df['salary'].apply(bagi10) 
df

end = time.time()

print(end-start)
```

**apply(lambda)**

```python
# series.apply(lambda_function)
df['salary'].apply(lambda gaji: gaji//10) 
```

```python
df['bonus_lambda']= df['salary'].apply(lambda gaji: gaji//10)
df
```

**Soal:** 
- Buat kolom baru bernama 'kelas_gaji'
- jika salary di atas 7000, maka kolom 'kelas_gaji' terisi dengan 'high'
- jika salary sama dengan atau di bawah 7000, maka kolom 'kelas_gaji' terisi dengan 'low'

```python
# cara 1: regular function
def kelas(gaji):
    if gaji > 7000:
        hasil = 'high'
    else:
        hasil = 'low'

    return hasil

kelas(7000)
```

```python
df['kelas_gaji'] = df['salary'].apply(kelas)
df
```

```python
# cara 2: lambda

df['kelas_gaji_lambda'] = df['salary'].apply(lambda x: 'high' if x>7000 else 'low')
df 
```

## Pivot Table

```python
df_sales = pd.DataFrame({
    'Kota': ['Jakarta', 'Jakarta','Jakarta','Tangerang','Tangerang','Tangerang'],
    'Cabang': [1,1,2,2,1,1],
    'Salesman': ['Jack','Alex','Jack','Alex','Jack','Alex'],
    'Profit': [10,20,30,40,50,60]
})

df_sales
```

```python
# soal: tampilkan total profit untuk tiap kota dari tiap salesman 
df_sales.groupby(by=['Kota','Salesman']).sum()[['Profit']]
```

```python
# pivot_table
df_sales.pivot_table(
    index='Kota',
    columns='Salesman',
    values='Profit',
    aggfunc= np.sum  # defaultnya aggfunc='mean'
)
```

```python
# pivot_table kemudian diurutkan
df_sales.pivot_table(
    index='Kota',
    columns='Salesman',
    values='Profit',
    aggfunc= np.sum 
).sort_values(by='Jack', ascending=False)
```
