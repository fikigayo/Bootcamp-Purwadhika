Measure of spread price

```python
range_price = df_car['price'].max() - df_car['price'].min() #range
std_price = df_car['price'].std()
iqr = df_car['price'].quantile(.75) - df_car['price'].quantile(.25)

print(f'range price : {range_price}')
print(f'std price : {std_price}')
print(f'iqr price : {iqr}')
```

outlayer : 1,5*iqr

```python
batas_bawah = df_car['price'].quantile(.25)-(1.5*iqr)
batas_atas = df_car['price'].quantile(.75)+(1.5*iqr)

print(f'outlier1 : {batas_bawah}')
print(f'outlier1 : {batas_atas}')
```

```python
# df_car[df_car['price'] > batas_atas]
```

```python
def outlier(dframe, column_):
    q1 = dframe[column_].quantile(0.25)
    q3 = dframe[column_].quantile(0.75)

    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr

    return dframe[(dframe[column_] < lower) | (dframe[column_] > upper)]
```

```python
outlier(df_car, 'price')
```
