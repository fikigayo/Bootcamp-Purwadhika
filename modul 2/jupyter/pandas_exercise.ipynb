## Create Dataframe üî≥

### Question 1

Membuat dataframe dengan 3 kolom yaitu x,y dan z. memiliki data berurutan dari setiap kolom angka 1 sampai 15

**Answer** ‚¨áÔ∏è

```python
data = {
    'x': list(range(1, 6)),
    'y': list(range(6, 11)),
    'z': list(range(11, 16))
}
df = pd.DataFrame(data)
df
```

### Question 2
Menambahkan kolom 'v' yang memiliki nilai data kebawah yaitu 16-20

```python
df['v'] = [x for x in range(16,21)]
df
```

**Answer** ‚¨áÔ∏è

## Data Manipulation üßô

### Question 1
Buat kolom baru nama 'f' yaitu dengan menggunakan nilai dari kolom 'v' apakah lebih besar dari 18 atau tidak

**Answer** ‚¨áÔ∏è

```python
lst = []
for i in df['v']:
    if i > 18:
        temp = 'High'
    else:
        temp = 'Low'
    lst.append(temp)
df['f'] = lst
df
```

### Question 2
Hapus kolom 'y'

**Answer** ‚¨áÔ∏è

```python
df.drop('y', axix='columns', inplace=True)
df
```

### Question 3
Urutkan data berdasarkan kolom 'v' secara descending

**Answer** ‚¨áÔ∏è

```python
df.sort_values('v',ascending=False)
df
```
