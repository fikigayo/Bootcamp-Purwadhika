# REGEX EXERCISE

```python
import re
```

```python
with open('mbox-short.txt', 'r') as infile:
    text = infile.read()
```

### 1. Tampilkan alamat email

Contoh: 
- 'stephen.marquard@uct.ac.za'
- 'postmaster@collab.sakaiproject.org'
- '200801051412.m05ECIaH010327@nakamura.uits.iupui.edu'
- 'source@collab.sakaiproject.org'

Ada 305 

```python
pattern1 = '[\w.]+@[\w.]+\.[a-zA-Z]{2,}'
re.findall(pattern=pattern1, string=text)
```

### 2. Tampilkan tanggal dan jam

Contoh
- '2008-01-05 09:12:07'
- '2008-01-04 18:08:50'
- '2008-01-04 16:09:01'

Ada 34

```python
pattern1 = '\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}'
re.findall(pattern=pattern1, string=text)
```

### 3. Tampilkan Hari dan tanggal

Contoh:
- 'Sat, 05 Jan 2008'
- 'Sat, 5 Jan 2008'
- 'Sat, 05 Jan 2008'

Ada 223

```python
pattern1 = '[A-Z][a-z]{2}, \d{1,2} [A-Z][a-z]{2} \d{4}'
re.findall(pattern=pattern1, string=text)
```

### 4. Tampilkan email yang didahului dengan Author

Contoh:
- 'stephen.marquard@uct.ac.za',
- 'louis@media.berkeley.edu',
- 'zqian@umich.edu'

Ada 27 email

```python
pattern1 = 'Author: (\S+@\S+)'
re.findall(pattern=pattern1, string=text)
```

### 5. Tampilkan email yang didahului dengan From atau To

Contoh:
- 'source@collab.sakaiproject.org',
- 'stephen.marquard@uct.ac.za',
- 'source@collab.sakaiproject.org',
- 'louis@media.berkeley.edu'

Ada 54 email

```python
pattern1 = '(From|To): (\S+@\S+)'
re.findall(pattern=pattern1, string=text)
```
