```python
df.isna().sum()
```

```python
df.head()
```

```python
df.groupby(by='JobTitle').mean()[['BasePay']]

# syntax cara 1:
# 1. groupby berdasarkan kolom categorical ('JobTitle')
# 2. lakukan aggregate function (mean)
# 3. indexing kolom yg diperlukan ('BasePay')
```

```python
df.groupby('JobTitle')[['BasePay']].mean()

# syntax cara 2:
# 1. groupby berdasarkan kolom categorical ('JobTitle')
# 2. indexing kolom yg diperlukan ('BasePay')
# 3. lakukan aggregate function (mean)
```

## Merge 

Menggabungkan 2 dataframe atau lebih menggunakan kolom tertentu (key column)

```python
df_jawa = pd.DataFrame({
    'pic': list('ABCD'),
    'Jakarta': [1,2,3,4],
    'Surabaya': [5,6,7,8],
})

df_jawa
```

```python
df_sumatra = pd.DataFrame({
    'pic': list('ABCE'),
    'Medan': [10,20,30,40],
    'Palembang': [50,60,70,80],
})

df_sumatra
```

```python
display(df_jawa, df_sumatra)
```

```python
# menggabungkan df_jawa dan df_sumatra menggunakan kolom 'pic'
pd.merge(left=df_jawa, right=df_sumatra)

# defaultnya how='inner'
# defaultnya on= mencari kolom dengan nama yg sama
```

```python
# syntax cara 1
pd.merge(left=df_jawa, right=df_sumatra, on='pic', how='inner')
```

```python
# syntax cara 2
df_jawa.merge(df_sumatra, on='pic', how='inner')
```

```python
pd.merge(left=df_jawa, right=df_sumatra, on='pic', how='outer', indicator=True)

# indicator=True untuk mengetahui pic (value pada key column) berasal dari table yg mana
```

```python
# ubah nama kolom 'pic' jadi 'salesman' di df_sumatra
df_sumatra = df_sumatra.rename(columns={'pic':'salesman'})
df_sumatra
```

```python
# menggabungkan 2 dataframe menggunakan kolom dengan nama yg berbeda (tapi isinya mengacu pada nilai yg sama)
pd.merge(left=df_jawa, right=df_sumatra, left_on='pic', right_on='salesman')
```

## Join

- join() mirip dengan merge()
- merge() menggabungkan 2 dataframe menggunakan kolom tertentu (key column)
- join() menggabungkan 2 dataframe menggunakan index
- defaultnya left join


```python
jawa = df_jawa.set_index('pic')
jawa.index.name = None
jawa
```

```python
sumatra = df_sumatra.set_index('salesman')
sumatra.index.name = None
sumatra 
```

```python
# menggabungkan dataframe jawa dan sumatra menggunakan index-nya

# syntax cara 1
jawa.join(sumatra)
```

```python
# syntax cara 2
pd.DataFrame.join(jawa, sumatra)
```

```python
jawa.join(sumatra, how='inner')
```

```python
# reset index
jawa1 = jawa.reset_index()
sumatra1 = sumatra.reset_index()

display(jawa1, sumatra1)
```

```python
jawa1.join(sumatra1, lsuffix='_kiri', rsuffix='_kanan', how='left')

# kalau ada kolom yg namanya sama, gunakan suffix
```
